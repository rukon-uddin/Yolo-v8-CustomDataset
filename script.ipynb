```python
from ultralytics import YOLO
model = YOLO("yolov8n.yaml")  # build a new model from scratch
model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)
```

```python
model.train(data="data.yaml", epochs=3)
```

```python
results = model("https://ultralytics.com/images/bus.jpg")
```

```python
for result in results:
    print(result.boxes.cls)
```

```python
from tqdm import tqdm
import os
import time

start = time.time() 
image_files = []
for filename in tqdm(os.listdir("data/obj/train")):   # prive the appropriate path
    if filename.endswith(".jpg") or filename.endswith(".JPG") or filename.endswith(".png") or filename.endswith(".PNG"):
        image_files.append("data/obj/train/" + filename) # prive the appropriate path (Note: dont forget to put he backslash '/' at the end)
os.chdir("data")
with open("train.txt", "w") as outfile:
    for image in tqdm(image_files):
        outfile.write(image)
        outfile.write("\n")
    outfile.close()
os.chdir("..")

print(f'\nTime: {time.time() - start}')
```
